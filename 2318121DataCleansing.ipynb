# 3. Menampilkan struktur variable data
Tampilkan struktur → cek isi data: tipe kolom, jumlah baris, contoh datanya.

```python
print("Struktur Data:")
print(df.info())
print("\nContoh 5 baris data awal:")
print(df.head())
```

# 4. Data Cleansing - Standarisasi kolom
Standarisasi nama kolom → biar konsisten (huruf kecil semua, tanpa spasi), mudah dipanggil di coding.

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(".", "_", regex=False)
```

# 5. Data Cleansing - Standarisasi kolom Nama.Lengkap
Standarisasi Nama.Lengkap → rapikan teks (hapus spasi berlebih, huruf jadi Title Case).

```python
df['nama_lengkap'] = df['nama_lengkap'].astype(str).str.strip().str.title()
```

# 6. Data Cleansing - Standarisasi kolom Tanggal.Lahir
Standarisasi Tanggal.Lahir → ubah semua ke format tanggal yang valid agar bisa dihitung.

```python
df['tanggal_lahir'] = pd.to_datetime(df['tanggal_lahir'], errors="coerce")
```

# 7. Data Cleansing - Deduplikasi Data
Deduplikasi data → hapus data ganda supaya tidak dobel saat analisis.

```python
df = df.drop_duplicates()
```

# 8. Data Enrichment (contoh: hitung umur dari tanggal lahir)
Enrichment (Umur) → tambahkan informasi baru dari data yang ada, misalnya umur dari tanggal lahir.
tujuan keseluruhan = membersihkan data supaya rapi, konsisten, valid, dan siap dianalisis.

```python
today = pd.to_datetime("today")
df['umur'] = df['tanggal_lahir'].apply(lambda x: today.year - x.year if pd.notnull(x) else np.nan)

print("\nData setelah cleansing dan enrichment:")
print(df.head())
```
